```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0,2.0])
y_train = np.array([300.0,500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
m = x_train.shape[0]
print(f"m = {m}")
```

```python
for i in range(m):
    print(f"x_{i} = {x_train[i]}")
    print(f"y_{i} = {y_train[i]}")
```

```python
plt.plot(x_train, y_train, 'x', c='r')
plt.title('House Price ')
plt.xlabel('Size in Square Feet')
plt.ylabel('Price in 1000 $')
plt.show()
```

```python
#  Set w and b to 100
w = 100
b = 100
print(f"w = {w}")
print(f"b = {b}")
```

```python
def forward_propagation(x_train, w, b):
    m = x_train.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x_train[i] + b
    return f_wb
```

```python
tmp_wb = forward_propagation(x_train, w, b)
plt.plot(x_train, tmp_wb, label='Our Prediction', c='b')
plt.scatter(x_train, y_train, label='Actual Prediction', c='r', marker='x')
plt.title('House Price ')
plt.xlabel('Size in Square Feet')
plt.ylabel('Price in 1000 $')
plt.legend()
plt.show()
```

```python
#  Set w= 200 and b = 100
w = 200
b = 100
tmp_wb = forward_propagation(x_train, w, b)
plt.plot(x_train, tmp_wb, label='Our Prediction', c='b')
plt.scatter(x_train, y_train, label='Actual Prediction', c='r', marker='x')
plt.title('House Price ')
plt.xlabel('Size in Square Feet')
plt.ylabel('Price in 1000 $')
plt.legend()
plt.show()
```

```python
def compute_cost(x,y,w,b):
    m = x.shape[0]
    total_cost = 0
    for i in range(m):
        f_wb = w * x[i] + b
        total_cost += (f_wb - y[i])**2
    total_cost = (1/(2*m)) * total_cost
    return total_cost
```
